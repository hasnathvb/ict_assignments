1 fizz buzz

```python
for i in range(1,101):
  if i%3==0 and 1%5==0:
    print('fizzbuzz')
  elif i%3==0:
    print(i,'=fizz')
  elif i%5==0:
    print(i,'=buzz')
  else:
    print(i)
```

2 electricity bill

```python
units = int(input("enter the number of units: "))
bill = 0
if units <= 100:
  bill = units + 5
elif units <= 200:
  bill = (100 * 5) + (units - 100) * 7
else:
  bill = (100 * 5) + (100 * 7) + (units - 200) * 10
print(f"total bill amount:{bill} ")
```
