```python
# Student Performance Analyzer

students = []

subjects = ["Maths", "Science", "English"]

# Add student details
def add_student():
    roll = int(input("Enter Roll Number: "))
    name = input("Enter Name: ")
    students.append({
        "roll": roll,
        "name": name,
        "marks": {},
        "average": 0,
        "status": ""
    })
    print("Student added successfully!\n")

# Enter marks with validation
def enter_marks():
    for student in students:
        print(f"\nEntering marks for {student['name']}")
        for sub in subjects:
            while True:
                mark = int(input(f"Enter marks for {sub}: "))
                if 0 <= mark <= 100:
                    student["marks"][sub] = mark
                    break
                else:
                    print("Invalid marks! Enter between 0 and 100.")

# Calculate average
def calculate_average(student):
    total = 0
    for mark in student["marks"].values():
        total += mark
    student["average"] = total / len(subjects)

# Grade rules
def get_grade(subject, marks):
    if subject == "Maths":
        if marks >= 90: return "A+"
        elif marks >= 75: return "A"
        elif marks >= 60: return "B"
        elif marks >= 40: return "C"
        else: return "F"

    elif subject == "Science":
        if marks >= 85: return "A+"
        elif marks >= 70: return "A"
        elif marks >= 55: return "B"
        elif marks >= 40: return "C"
        else: return "F"

    elif subject == "English":
        if marks >= 80: return "A+"
        elif marks >= 65: return "A"
        elif marks >= 50: return "B"
        elif marks >= 40: return "C"
        else: return "F"

# Pass or fail status
def pass_fail_status(student):
    for mark in student["marks"].values():
        if mark < 40:
            student["status"] = "Fail"
            return
    student["status"] = "Pass"

# Subject-wise toppers
def subject_toppers():
    for sub in subjects:
        top_marks = -1
        topper = ""
        for student in students:
            if student["marks"][sub] > top_marks:
                top_marks = student["marks"][sub]
                topper = student["name"]
        print(f"Topper in {sub}: {topper} ({top_marks})")

# Ranking without built-in sorting
def rank_students():
    ranked = students[:]
    n = len(ranked)
    for i in range(n):
        for j in range(i + 1, n):
            if ranked[i]["average"] < ranked[j]["average"]:
                ranked[i], ranked[j] = ranked[j], ranked[i]
    return ranked

# Display report
def display_report():
    for student in students:
        calculate_average(student)
        pass_fail_status(student)

    print("\n--- STUDENT REPORT ---")
    for student in students:
        print(f"\nRoll No: {student['roll']}")
        print(f"Name: {student['name']}")
        for sub, mark in student["marks"].items():
            grade = get_grade(sub, mark)
            print(f"{sub}: {mark} | Grade: {grade}")
        print(f"Average: {student['average']:.2f}")
        print(f"Status: {student['status']}")

    print("\n--- SUBJECT TOPPERS ---")
    subject_toppers()

    print("\n--- OVERALL RANKING ---")
    ranked = rank_students()
    rank = 1
    for student in ranked:
        print(f"Rank {rank}: {student['name']} ({student['average']:.2f})")
        rank += 1

# Menu-driven system
def menu():
    while True:
        print("\n1. Add Student")
        print("2. Enter Marks")
        print("3. Generate Report")
        print("4. Exit")
        choice = input("Enter your choice: ")

        if choice == "1":
            add_student()
        elif choice == "2":
            enter_marks()
        elif choice == "3":
            display_report()
        elif choice == "4":
            print("Exiting Program...")
            break
        else:
            print("Invalid choice! Try again.")

menu()
```

```python
# Quiz Engine with Dynamic Scoring

quiz = {
    "Easy": [
        {"q": "What is the capital of India?",
         "options": ["A. Delhi", "B. Mumbai", "C. Chennai", "D. Kolkata"],
         "ans": "A"},

        {"q": "Which is the largest planet?",
         "options": ["A. Earth", "B. Mars", "C. Jupiter", "D. Venus"],
         "ans": "C"}
    ],

    "Medium": [
        {"q": "Who wrote 'Wings of Fire'?",
         "options": ["A. Nehru", "B. APJ Abdul Kalam", "C. Tagore", "D. Gandhi"],
         "ans": "B"},

        {"q": "Which data structure uses FIFO?",
         "options": ["A. Stack", "B. Tree", "C. Queue", "D. Graph"],
         "ans": "C"}
    ],

    "Hard": [
        {"q": "What is the time complexity of binary search?",
         "options": ["A. O(n)", "B. O(log n)", "C. O(n^2)", "D. O(1)"],
         "ans": "B"},

        {"q": "Which keyword is used for exception handling?",
         "options": ["A. try", "B. error", "C. catch", "D. handle"],
         "ans": "A"}
    ]
}

scores = {"Easy": 1, "Medium": 2, "Hard": 3}
negative = {"Easy": 0.25, "Medium": 0.5, "Hard": 1}

def attempt_quiz():
    name = input("Enter participant name: ")
    total_score = 0
    correct = 0
    wrong = 0
    attempt_log = []

    seed = len(name)  # used for pseudo-random logic

    print("\n--- QUIZ STARTED ---\n")

    for level in quiz:
        questions = quiz[level]
        index = seed % len(questions)

        for _ in range(len(questions)):
            q = questions[index]

            print(f"[{level}] {q['q']}")
            for opt in q["options"]:
                print(opt)

            answer = input("Your answer (A/B/C/D): ").upper()

            if answer == q["ans"]:
                total_score += scores[level]
                correct += 1
                result = "Correct"
            else:
                total_score -= negative[level]
                wrong += 1
                result = "Wrong"

            attempt_log.append(
                f"{level} | {q['q']} | Your Answer: {answer} | {result}"
            )

            index = (index + seed) % len(questions)

    generate_report(name, total_score, correct, wrong, attempt_log)

def generate_report(name, score, correct, wrong, log):
    print("\n--- QUIZ REPORT ---")
    print(f"Participant: {name}")
    print(f"Total Score: {score}")
    print(f"Correct Answers: {correct}")
    print(f"Wrong Answers: {wrong}")

    print("\n--- DETAILED LOG ---")
    for entry in log:
        print(entry)

def menu():
    while True:
        print("\n1. Start Quiz")
        print("2. Exit")
        choice = input("Enter choice: ")

        if choice == "1":
            attempt_quiz()
        elif choice == "2":
            print("Thank you for participating!")
            break
        else:
            print("Invalid choice! Try again.")

menu()
```
